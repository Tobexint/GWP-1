```python
#import pandas_ta_classic as ta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
import ta
import yfinance as yf
import time
```

```python
xc = pd.read_csv("ECH-adj-data.csv")
xc.head()
```

```python
xc = xc.iloc[2:].copy()
```

```python
xc.rename(columns={"Price": "Date"}, inplace=True)
```

```python
xc["Date"] = pd.to_datetime(xc["Date"])
```

```python
cols = ["Close", "High", "Low", "Open", "Volume"]

xc[cols] = xc[cols].apply(pd.to_numeric)
```

```python
xc.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

plt.plot(xc["Date"], xc["Close"], linewidth=1.5)

plt.title("Figure 1: ECH Historical Closing Price (2009–2020)")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
